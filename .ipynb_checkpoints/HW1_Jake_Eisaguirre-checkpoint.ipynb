```python
df['datetime'] = pd.to_datetime(df['time'], unit = 'ms')
print(df)
```

```python
plt.figure(figsize = (10, 6), dpi = 300)
plt.plot(df['datetime'], df['prcp'])
plt.title('Santa Barbara Rainfall', fontsize = 16)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Precip (mm)', fontsize = 14)
```

```python
ldat = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA')
```

```python
ldat_time_filter = ldat.filter(ee.Filter.date('2013-04-11', '2021-09-28'))
```

```python
sb = ee.Geometry.Point([-119.69, 34.42]) 
```

```python
ldat_filtered = ldat_time_filter.filterBounds(sb)
```

```python
ldat_little_cloud = ldat_filtered.sort('CLOUD_COVER').first()
```

```python
parameters = {'bands': ['B4', 'B3', 'B2'],
             'min': 0.0,
             'max': 0.3
            }
```

```python
lmap = geemap.Map(center=[34.43, -119.69], zoom=12)
```

```python
lmap.addLayer(ldat_little_cloud, parameters)
```

```python
red = ldat_little_cloud.select('B4')
nir = ldat_little_cloud.select('B5')
```

```python
ndvi=(nir.subtract(red)).divide((nir.add(red))).rename('NDVI')
```

```python
ndviParams = {'min': -1, 
              'max': 1, 
              'palette': ['blue', 'white', 'green']
             }

lmap.addLayer(ndvi, ndviParams,'NDVI')

lmap
```

```python
rain_dat = ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA')
```

```python
rain_dat_time_filter = rain_dat.filter(ee.Filter.date('2020-12-26', '2020-12-30'))
```

```python
sb = ee.Geometry.Point([-119.69, 34.42]) 
```

```python
rain_dat_filtered = rain_dat_time_filter.filterBounds(sb)
```

```python
rain_dat_little_cloud = rain_dat_filtered.sort('CLOUD_COVER').first()
```

```python
parameters = {'bands': ['B4', 'B3', 'B2'],
             'min': 0.0,
             'max': 0.3
            }
```

```python
rain_map = geemap.Map(center=[34.42, -119.69], zoom=12)
```

```python
red = ldat_little_cloud.select('B4')
nir = ldat_little_cloud.select('B5')
```

```python
rain_ndvi=(nir.subtract(red)).divide((nir.add(red))).rename('RAIN_NDVI')
```

```python
ndviParams = {'min': -1, 
              'max': 1, 
              'palette': ['blue', 'white', 'green']
             }
```

```python
rain_map.addLayer(rain_ndvi, ndviParams,'RAIN_NDVI')

rain_map
```
